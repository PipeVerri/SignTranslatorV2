```python
# Analizar las probabilidades de rnn_scores para decidirme en un threshold
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv("rnn_scores.csv")
print(df.head())
```

```python
# Graficar el score promedio de cada se√±a
mean_scores = df.groupby("label")["softmax_score"].mean()
plt.figure(figsize=(12, 5))
mean_scores.plot(kind="bar")
```
